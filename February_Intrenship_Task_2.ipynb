**1.Unique Words in a Sentence**

```python
sentence=input("Enter Sentence:")
#splitting the words using the spaces
words=sentence.split()
#convert the words list into set to get unique words
unique_word=set(words)
print("Unique Words Count:" ,len(unique_word))
print("Unique Words:", unique_word)
```

```python
sentence=input("Enter Sentence:")
#splitting the words using the spaces
words=sentence.split()
unique=[]
for word in words:
  if word not in unique:
    unique.append(word)
print("Unique Words Count:" ,len(unique))
print("Unique Words:", unique)
```

**2.Highest Salary from Employee Data**

```python
#Employee dictionary having employee names and their salaries
employees={"Ravi":75000,"Anita":68000,"Kiran":72000}
#Initalize highest salary to 0
highest_salary=0
highest_employee=""
for name in employees:
  if employees[name]>highest_salary:
    highest_salary=employees[name]
    highest_employee=name
print(f"Highest Salary: {highest_employee} - {highest_salary}")
```

**3.Find Maximum and Minimum Values**

```python
#numbers List
numbers = [45, 22, 89, 10, 66]
#Initalize maximum and minimum values with the first element of the list
max_value=numbers[0]
min_value=numbers[0]
for n in numbers:
  if n>max_value:
    max_value=n
  if n<min_value:
    min_value=n
print("List: ",numbers)
print("Max: ",max_value)
print("Min: ",min_value)
```

**4.Count Products Above a Price Threshold**

```python
#price of a products list
prices = [450, 1200, 899, 1500, 300]
count=0
for price in prices:
  if price>1000:
    count+=1
print("Products above 1000: ",count)
```

**5.Calculate Attendance Percentage**

```python
#Attendance List
attendance = ["P", "P", "A", "P", "P"]
#count total days and present days
total=len(attendance)
present=0

for day in attendance:
  if day=="P":
    present+=1
#Calculate attendance percentage
attendance_percentage=(present/total)*100
print("Attendance Percentage: ",attendance_percentage)
```

**7.Count Character Frequency**

```python
text=input()
char_frequency={}
for char in text:
  if char in char_frequency:
    char_frequency[char]+=1
  else:
    char_frequency[char]=1
print(char_frequency)
```
